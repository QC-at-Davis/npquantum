### X Gate

```python
new_state = X @ ket0 # ket0 * X
print(new_state)
plot_bloch_multivector(to_qiskit_plot(new_state))
```

```python
new_state = X @ ket1 # ket0 * X
print(new_state)
plot_bloch_multivector(to_qiskit_plot(new_state))
```

### Y Gate

```python
new_state = Y @ ket0
print(new_state)
plot_bloch_multivector(to_qiskit_plot(new_state))
```

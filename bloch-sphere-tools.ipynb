```python
# |psi> = cos(theta/2)|0> + e^(i*phi)*sin(theta/2)|1>
# 0 <= theta <= pi
# 0 <= phi <= 2*pi
```

```python
theta = -np.pi/3
```

```python
phi = np.pi/4
```

```python
coeff = []
coeff.append(np.cos(theta/2))
coeff.append((np.e**(1j*phi))*np.sin(theta/2))
state_vec = coeff[0]*ket0 + coeff[1]*ket1
```

```python
state_vec
```

```python
state_vec.H * state_vec
```

```python
plot_bloch_multivector(to_qiskit_plot(state_vec))
```
